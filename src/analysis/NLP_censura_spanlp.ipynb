### Censurar insultos

```python
from spanlp.palabrota import Palabrota
palabrota = Palabrota()
print(palabrota.contains_palabrota("Sapo"))
print(palabrota.censor("Hola huevon cómo está?"))
palabrota = Palabrota(censor_char="*")
print(palabrota.censor("Hola marica cómo está?"))
# salida: True
```

```python
from spanlp.palabrota import Palabrota
from spanlp.domain.countries import Country

palabrota = Palabrota(censor_char="*", countries=[Country.COLOMBIA], include=["mondá"])
print(palabrota.censor("Hola monda marica cómo está?"))

# salida: Hola @@@@@@ @@@@@@ cómo está?
```

```python
import numpy as np
# Define a function to apply row-wise
def check_palabrota(row):
    if row["depth"] == 0.0:
        text = row["commentText"]
    elif pd.isna(row["depth"]):
        text = row["postText"]
    else:
        return False  # or np.nan if you prefer
    return palabrota.contains_palabrota(str(text))  # ensure it's string
def censor_palabrota(row):
    if row["has_palabrota"]:
        if row["depth"] == 0.0:
            text = row["commentText"]
        elif pd.isna(row["depth"]):
            text = row["postText"]
        else:
            return np.nan
        return palabrota.censor(str(text))
    else:
        return np.nan
```

```python
df_clean.loc[774]
```

```python
df_clean["has_palabrota"] = df_clean.apply(check_palabrota, axis=1)
```

```python
df_clean["censored"] = df_clean.apply(censor_palabrota, axis=1)
```

```python

palabrota_rows = df_clean[df_clean["has_palabrota"] == True]
print("Número de entradas: " + str(df_clean.shape[0]))
print("Entradas identificadas con palabrota: " + str(palabrota_rows.shape[0]))
print(palabrota_rows[["postText", "commentText", "depth", "censored"]].sample())
```

### Análisis sentimientos


```python
from transformers import pipeline

# Load the classification pipeline with the specified model
pipe = pipeline("text-classification", model="tabularisai/multilingual-sentiment-analysis")

# Classify a new sentence
sentence = "Buenas noches gente, pasó por acá para que por favor me ayuden a difundir esta imagen de mi gato que lamentablemente desapareció hace tres días en el Barrio La Independencia, el barrio queda detrás de la uis por los lados de la entrada de la 25 en dónde le conocen como 'las escaleras'. El gato tiene 5 años y ya está operado, se llama Chiki y tiene una pequeña cortadura en una oreja de cuando se peleó con otro gato. Si alguien conoce alguna otra información se lo agradezco"
result = pipe(sentence)

# Print the result
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelWithLMHead

tokenizer = AutoTokenizer.from_pretrained("mrm8488/t5-base-finetuned-emotion")

model = AutoModelWithLMHead.from_pretrained("mrm8488/t5-base-finetuned-emotion")

def get_emotion(text):
  input_ids = tokenizer.encode(text + '</s>', return_tensors='pt')

  output = model.generate(input_ids=input_ids,
               max_length=2)

  dec = [tokenizer.decode(ids) for ids in output]
  label = dec[0]
  return label

get_emotion("buenas noches gente, pasó por acá para que por favor me ayuden a difundir esta imagen de mi gato que lamentablemente desapareció hace tres días en el Barrio La Independencia, el barrio queda detrás de la uis por los lados de la entrada de la 25 en dónde le conocen como 'las escaleras'. El gato tiene 5 años y ya está operado, se llama Chiki y tiene una pequeña cortadura en una oreja de cuando se peleó con otro gato. Si alguien conoce alguna otra información se lo agradezco")
```
